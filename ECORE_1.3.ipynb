```python
#Importação da biblioteca Pandas como 'pd' para utilização otimizada no código;
import pandas as pd
```

```python
#Leitura de CSV com ajustes para separar os dados em colunas considerando multiplos delimiters;
df = pd.read_csv(
    'movies_ecore.csv',
    sep=';|,', 
    engine='python', 
    names=['Number', 'Movie', 'Year']
   )
#remoção de parenteses dos dados para separação de ano e nome do filme em colunas diferentes;
df = df.replace(to_replace='\)', value="", regex=True)
df = df.replace(to_replace='\(', value="", regex=True)
df.set_index('Number')
```

```python
#Agrupamento e contagem de filmes por ano de lançamento
df1 = df.groupby(['Year']).size().reset_index(name='Counts')

#Separação dos 9 anos com menos lançamentos de filmes
df1 = ((df1.sort_values('Counts', ascending = True)).set_index('Year')).head(9)
#Executando o comando abaixo, a tela irá mostrar a lista dos 9 anos com menos lançamentos de filmes
#em ordem cescente
df1
```
