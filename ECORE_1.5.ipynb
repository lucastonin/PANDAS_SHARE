```python
#Importação da biblioteca Pandas como 'pd' para utilização otimizada no código;
import pandas as pd
```

```python
#Leitura do arquivo CSV da atividade, criando uma DataFrame utilizando 'DELIMITER' para sinalizar para o programa qual 
#caracter de separação deve ser considerado;
la = pd.read_csv('customers_rating_ecore.csv', delimiter = ';')

#Identificação das avaliações feitas por data;
la = la.sort_values(by = ['Date'])

#Criação de máscara para valor máximo da coluna 'Date'(última data registrada na lista);
mask1 = la['Date'].max()

#Criação de DataFrame apenas com avaliações feitas na última data registrada na lista e Movie Id;
lb = la.loc[la['Date'] == mask1]

#Criação de DataFrame com nota média de avaliação dos filmes na data selecionada;
lb = lb.groupby(['Movie_Id'])['Rating'].mean()
lb = lb.to_frame()
#DataFrame com nota média de avaliação dentro da data específica maiores que 4.7, conforme pedido do cliente;
fl = lb.loc[lb['Rating'] >= 4.7]

#DataFrame com duas colunas: Movie_Id e Rating;
fl = fl[['Rating']]
fl
```

```python
#DataFrame com os 10 filmes com menores médias de avaliação dentro do dos requisitos acima;
bttm10 = (fl.sort_values('Rating', ascending = True)).head(10)
```

```python
#Leitura de CSV com nome dos filmes 
names = (pd.read_csv('movies_ecore.csv',  delimiter = ';' , names = ['Movie_Id', 'Movie'])).set_index('Movie_Id')

#Juntando listas com avaliações médias com a de nomes dos filmes, com  a coluna de Movie_Id de intersecção;
finallist = (bttm10.merge(names, on = ['Movie_Id']))
```

```python
#Exposição de dados solicitados pelo cliente;
print('Segue abaixo a lista dos 10 piores filmes com nota (média) acima de 4.7 e suas respectivas notas ')
finallist = finallist.sort_values('Rating' , ascending = False)
finallist.set_index('Movie')
```
