```python
#Importação da biblioteca Pandas como 'pd' para utilização otimizada no código;
import pandas as pd
```

```python
#Leitura de CSV atribuindo delimiter específico
df = pd.read_csv('customers_rating_ecore.csv',  delimiter = ';')
```

```python
#DataFrame apenas com Rating e Movie_Id's
df1= df[['Rating', 'Movie_Id']]

#Agrupamento de avaliações feitas por filme,e cálculo da nota média de avaliação de cada filme 
df1 = df.groupby(['Movie_Id'])['Rating'].mean()
df1 = df1.to_frame()
df1
```

```python
#Selecionando as 5 maiores avaliações no DataFrame com Ratings decrescentes
df3 = df1.sort_values('Rating', ascending=False)
df3
top5 = df.head(5)
top5
```

```python
#Leitura do DataFrame secundário para obter nomes dos filmes
names = pd.read_csv('movies_ecore.csv',  delimiter = ';' , names = ['Movie_Id', 'Movie'])
names = names.set_index('Movie_Id')
names
```

```python
#Merging das duas DataFrames para atrelar o nome do filme à avaliação
common = top5.merge(names, on=['Movie_Id'])

#Comando que mostra a tabela com 5 filmes com maiores avaliações médias
common
```
